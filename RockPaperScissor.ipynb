```python
import tkinter as tk
from tkinter import PhotoImage
from PIL import Image, ImageTk
import random

# Game options
OPTIONS = ["Rock", "Paper", "Scissors"]

class RockPaperScissors:
    def __init__(self, root):
        self.root = root
        self.root.title("Rock Paper Scissors")
        self.root.configure(bg="#ADD8E6")  # Light blue background
        
        self.user_score = 0
        self.computer_score = 0
        
        # Labels
        self.label = tk.Label(root, text="Choose Rock, Paper, or Scissors", font=("Arial", 14), bg="#ADD8E6")
        self.label.pack(pady=10)
        
        # Load JPG images
        # Load JPG images and store as instance variables
        self.rock_img = Image.open("C:\\Users\\yuvay\\Downloads\\rock.jpg").resize((100, 100))
        self.paper_img = Image.open("C:\\Users\\yuvay\\Downloads\\paper.jpg").resize((100, 100))
        self.scissors_img = Image.open("C:\\Users\\yuvay\\Downloads\\scissors.jpg").resize((100, 100))

        # Convert to PhotoImage format (store in self to prevent garbage collection)
        self.rock_tk = ImageTk.PhotoImage(self.rock_img)
        self.paper_tk = ImageTk.PhotoImage(self.paper_img)
        self.scissors_tk = ImageTk.PhotoImage(self.scissors_img)

        # Store in a dictionary for easy reference
        self.image_dict = {"Rock": self.rock_tk, "Paper": self.paper_tk, "Scissors": self.scissors_tk}
        
        # Buttons with images
        self.buttons_frame = tk.Frame(root, bg="#ADD8E6")
        self.buttons_frame.pack()
        
        for option in OPTIONS:
            btn = tk.Button(self.buttons_frame, image=self.image_dict[option], command=lambda opt=option: self.play(opt), bg="#87CEEB", borderwidth=3)
            btn.pack(side=tk.LEFT, padx=10, pady=10)
        
        # Result Display
        self.result_label = tk.Label(root, text="", font=("Arial", 12), bg="#ADD8E6")
        self.result_label.pack(pady=10)
        
        # Score Display
        self.score_label = tk.Label(root, text=f"User: {self.user_score}  Computer: {self.computer_score}", font=("Arial", 12), bg="#ADD8E6")
        self.score_label.pack(pady=10)
        
        # Play Again Button
        self.play_again_button = tk.Button(root, text="Play Again", font=("Arial", 12), command=self.reset_game, bg="#20B2AA", fg="white")
        self.play_again_button.pack(pady=5)
        
    def play(self, user_choice):
        computer_choice = random.choice(OPTIONS)
        result = self.determine_winner(user_choice, computer_choice)
        
        if result == "Win":
            self.user_score += 1
        elif result == "Lose":
            self.computer_score += 1
        
        self.result_label.config(text=f"You chose {user_choice}, Computer chose {computer_choice}. {result}!")
        self.score_label.config(text=f"User: {self.user_score}  Computer: {self.computer_score}")
        
    def determine_winner(self, user, computer):
        if user == computer:
            return "It's a Tie"
        elif (user == "Rock" and computer == "Scissors") or \
             (user == "Scissors" and computer == "Paper") or \
             (user == "Paper" and computer == "Rock"):
            return "Win"
        else:
            return "Lose"
        
    def reset_game(self):
        self.result_label.config(text="")
        self.score_label.config(text=f"User: {self.user_score}  Computer: {self.computer_score}")

if __name__ == "__main__":
    root = tk.Tk()
    app = RockPaperScissors(root)
    root.mainloop()
```
